```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://docs.google.com/spreadsheets/d/1fCM2LCJd-sakIMCBgxAi03KyTWVFCBxfqHE2BTSj-Ig/gviz/tq?tqx=out:csv&sheet=Sheet1"
attrdata = pd.read_csv(url)

# Display first few rows
print(attrdata.head())

# General Info
print("\n--- Data Info ---")
print(attrdata.info())

# Missing values
print("\n--- Missing Values ---")
print(attrdata.isnull().sum())
```

```python
# Clean string columns
attrdata['Gender'] = attrdata['Gender'].astype(str).str.strip().str.title()
attrdata['Tech/NonTech'] = attrdata['Tech/NonTech'].astype(str).str.strip().str.title()
attrdata['Job Type'] = attrdata['Job Type'].astype(str).str.strip().str.title()
attrdata['Payroll'] = attrdata['Payroll'].astype(str).str.strip().str.title()
attrdata['Hired Through'] = attrdata['Hired Through'].astype(str).str.strip().str.title()

```

```python
print(attrdata['Gender'].value_counts())

attrdata['Gender'].value_counts().plot(kind='bar', color=['lightblue', 'salmon'], title='Gender Distribution')
plt.ylabel('Count')
plt.show()
```

```python
attrdata['DOB'] = pd.to_datetime(attrdata['DOB'], errors='coerce')
attrdata['Age'] = (pd.to_datetime('today') - attrdata['DOB']).dt.days // 365

print(attrdata['Age'].describe())

sns.histplot(attrdata['Age'].dropna(), kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
attrdata['DOJ'] = pd.to_datetime(attrdata['DOJ'], errors='coerce')
attrdata['Tenure (Years)'] = (pd.to_datetime('today') - attrdata['DOJ']).dt.days // 365

print(attrdata['Tenure (Years)'].describe())

sns.histplot(attrdata['Tenure (Years)'].dropna(), kde=True, bins=20)
plt.title('Tenure Distribution')
plt.xlabel('Years at Company')
plt.ylabel('Frequency')
plt.show()
```

```python
attrdata['Tech/NonTech'].value_counts().plot(kind='bar', title='Tech vs Non-Tech', color='teal')
plt.ylabel('Count')
plt.show()
```

```python
attrdata['Designation'].value_counts().head(10).plot(kind='barh', title='Top 10 Designations', color='goldenrod')
plt.xlabel('Count')
plt.show()
```

```python
attrdata['Department'].value_counts().plot(kind='bar', title='Department Distribution', color='skyblue')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
attrdata['Job Type'].value_counts().plot(kind='bar', title='Job Type Distribution', color='violet')
plt.ylabel('Count')
plt.show()
```

```python
attrdata['Reporting Manager Name'].value_counts().head(10).plot(kind='barh', title='Top 10 Managers by Team Size')
plt.xlabel('Employees')
plt.show()
```

```python
attrdata['Payroll'].value_counts().plot(kind='bar', title='Payroll Category', color='green')
plt.ylabel('Count')
plt.show()
```

```python
attrdata['Hired Through'].value_counts().plot(kind='bar', title='Hired Through', color='gray')
plt.ylabel('Count')
plt.show()
```

```python
# Gender vs Job Type
pd.crosstab(attrdata['Gender'], attrdata['Job Type']).plot(kind='bar', stacked=True, title='Gender vs Job Type')
plt.ylabel('Count')
plt.show()

# Department vs Tech/NonTech
pd.crosstab(attrdata['Department'], attrdata['Tech/NonTech']).plot(kind='bar', stacked=True, figsize=(10,5), title='Department vs Tech/NonTech')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
